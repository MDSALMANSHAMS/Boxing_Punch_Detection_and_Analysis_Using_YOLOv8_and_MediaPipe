```python
# Train YOLOv8 model
model = YOLO("yolov8n.pt")  # Use "yolov8n.pt" for a smaller model
results = model.train(data="/content/drive/MyDrive/Assignments/Sponsorlytix_Assignment/dataset.yaml", epochs=100, imgsz=640)
```
